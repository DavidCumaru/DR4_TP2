```python
import yaml

def executar_prompt(prompt):
    chat = model.start_chat(history=[])
    response = chat.send_message(prompt)
    return response.text

with open("roteiro.yml", "r") as file:
    roteiro = yaml.safe_load(file)

resultados = {}
for cidade, prompt in roteiro["roteiro"].items():
    print(f"Executando prompt para: {cidade.capitalize()}")
    resultado = executar_prompt(prompt)
    resultados[cidade] = resultado
    print(f"Resultado para {cidade.capitalize()}:\n{resultado}\n")

with open("resultados_prompts.txt", "w") as file:
    for cidade, resultado in resultados.items():
        file.write(f"Resultado para {cidade.capitalize()}:\n{resultado}\n")
```
